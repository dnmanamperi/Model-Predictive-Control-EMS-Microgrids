```python
import pandas as pd
%pip install openpyxl
# Load your CSV file
df = pd.read_csv("boralesgamuwa_weather_2025_01_01_to_2025_12_31_15min_interpolated_power_added_final_set.csv")

# Select column G (7th column)
col = df.iloc[:, 6]

# Count how many times each value appears
counts = col.value_counts().reset_index()

# Rename columns for clarity
counts.columns = ["Value", "Count"]

# Save to Excel
counts.to_excel("value_counts_output.xlsx", index=False)

print("Done! File saved as value_counts_output.xlsx")
```
